# 61
“United States”と”U.S.”のコサイン類似度を計算せよ．

```python
import numpy as np
```

```python
def cos_sim(v1,v2):
    return np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
```

```python
cos_sim(model['United_States'],model['U.S.'])
#model.similarity('United_States','U.S.')
```

# 66. WordSimilarity-353での評価
The WordSimilarity-353 Test Collectionの評価データをダウンロードし，単語ベクトルにより計算される類似度のランキングと，人間の類似度判定のランキングの間のスピアマン相関係数を計算せよ．

combined.csv<br>
Word 1,Word 2,Human (mean)

love,sex,6.77

tiger,cat,7.35

tiger,tiger,10.00

book,paper,7.46

computer,keyboard,7.62

```python
ws353=[]
with open('combined.csv', 'r') as f:
    next(f)#word1,word2を飛ばす
    for line in f:
        list=[]
        line=line.split(',')
        for l in line:
            list.append(l.replace('\n',''))
        new_vec= model.similarity(line[0],line[1])
        list.append(new_vec)
        ws353.append(list)
```

```python
for i in range(5):
    print(ws353[i])
```

```python
#二つの変数間の順位の統計的依存性　今回はモジュールを使用
def spearman(word1,word2):
    word1=np.array(word1)
    word2=np.array(word2)
    N=len(word1)
    
    return 1-(6*sum((word1-word2)**2)/(N*(N**2-1)))
```

```python
import numpy as np
from scipy.stats import spearmanr
word1=np.array(ws353).T[2]
word2=np.array(ws353).T[3]
correlation,pvalue=spearmanr(word1,word2)
print(f'スピアマン相関係数:{correlation}')
#spearman(ws353[0].T)
```

# 67. k-meansクラスタリング
国名に関する単語ベクトルを抽出し，k-meansクラスタリングをクラスタ数k=5として実行せよ．

非階層型クラスタリングのアルゴリズム<br>
クラスタの平均を用いて、与えられたクラスタ数 k個に分類することからこの名前が付けられました。k平均法と呼ばれる

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
i=1
num=[]
countries=[]
with open ('questions-words.txt') as file:
    next(file)
    for f in file:
        f=f.split()
        if len(f)==4:
            if f[3] not in countries:
            #print(f[3])
                num.append(model[f[3]])
                countries.append(f[3])
            
        else:
            if f[1]=='currency':
                break;

        


#sc=preprocessing.StandardScaler()
#sc.fit(X)

#https://techacademy.jp/magazine/28780    

kmeans=KMeans(n_clusters=5,random_state=0)
kmeans.fit(num)
for i in range(5):
    cluster=np.where(kmeans.labels_==i)[0]
    print('cluster',i)
    print(', '.join([countries[k] for k in cluster]))
```

# 68. Ward法によるクラスタリングPermalink
国名に関する単語ベクトルに対し，Ward法による階層型クラスタリングを実行せよ．さらに，クラスタリング結果をデンドログラムとして可視化せよ

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage

plt.figure(figsize=(15,5))
Z=linkage(num,method='ward')
dendrogram(Z,labels=countries)#デンドログラム、Y軸はユーグリッド距離
plt.show
```
